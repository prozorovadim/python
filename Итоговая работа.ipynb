### 1.2 Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.hist(data['price'])
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
```

## 2.2 Изучите распределение этажей домов

```python
import seaborn as sns

sns.barplot(x=data['floors'].value_counts().index, y=data['floors'].value_counts())
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количесво домов');
```

### Вывод: Дома преимущественно одноэтажные и двухэтажные

## 2.2 Изучите распределение состояния домов

### Вывод: преимущественно среднее состояние

## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np

data = data.drop('date', axis=1)
corr_matrix = data.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix)<0.3] = 0
corr_matrix
```

```python
plt.figure(figsize = (10, 8))
sns.heatmap(corr_matrix, annot = True, linewidths = 5, cmap = 'coolwarm')
```

```python
sns.jointplot(x = data['sqft_living'], y = data['price'], kind = 'reg');
```

### Вывод: Чем больше жилплощадь дома, тем выше его стоимость

```python
sns.jointplot(x=data['grade'], y=data['price'], kind='reg');
```

### Вывод: Чем выше качество конструкции и дизайна дома, тем выше его стоимость

```python
plt.figure(figsize = (10, 8))

sns.boxplot(x = data['price'], y = data['grade'].astype('str'), whis = 1.5)
            
plt.title('Распределение цены в зависимости от качества постройки и дизайна')
plt.xlabel('Цена')
plt.ylabel('Качество постройки и дизайна');
```

### Вывод: Наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13

```python
plt.figure(figsize = (10, 8))

sns.boxplot(x = data['price'], y = data['view'].astype('str'), whis = 1.5)
            
plt.title('Распределение цены в зависимости от вида из окна')
plt.xlabel('Цена')
plt.ylabel('Вид из окна');### Вывод: 
```
